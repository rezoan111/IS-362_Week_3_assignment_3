```python
import pandas as pd

# my odometer totals (cumulative miles) each day
cumulative_miles_list = [55, 120, 180, 240]  
```

```python
cumulative = pd.Series(
    cumulative_miles_list,
    index=range(1, len(cumulative_miles_list) + 1),
    name="cumulative_miles"
)

daily = cumulative.diff()
daily.iloc[0] = cumulative.iloc[0]   # day 1 miles = first total
daily.name = "miles_today"

summary = pd.DataFrame({
    "cumulative_miles": cumulative,
    "miles_today": daily
})

summary
```

```python
print("Total miles:", int(cumulative.iloc[-1]))
```
